```python
df=df[['coaster_name','Location','Status','Manufacturer','year_introduced',
       'latitude','longitude','Type_Main','opening_date_clean','speed_mph',
       'height_ft','Inversions_clean','Gforce_clean']].copy()
```

```python
df['Opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df=df.rename(columns={'coaster_name':'Coaster_Name','year_introduced':'Year_introduced',
                      'opening_date_clean':'Opening_Date','speed_mph':'Speed_mph','height_ft':'Height_ft',
                     'Inversion_clean':'Inversions','Gforce_clean':'Gforce'})
```

```python
# df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])]
# → Filters the DataFrame to remove duplicate rows based on these 3 columns.
# → The '~' inverts the boolean result, so we **keep only the unique rows**.

# .reset_index(drop=True)
# → Resets the index after filtering so it starts from 0 and increments cleanly.
# → drop=True ensures the **old index is discarded**, not added as a new column.

# .copy()
# → Creates a new, independent copy of the filtered DataFrame.
# → This avoids unexpected behavior or warnings when making further edits.

# ✅ Final result: A fresh DataFrame with duplicates removed (based on specific columns),
# clean index, and safe to modify further.
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

```python
# Creating a bar graph showing the year in which maximum coaster were introduced
df['Year_introduced'].value_counts().head(10)
```
